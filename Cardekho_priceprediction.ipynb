Splitting the data set into Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

Import the models

```python
pip install xgboost
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
lr= LinearRegression()
lr.fit(X_train,Y_train)

las= Lasso()
las.fit(X_train,Y_train)

rf= RandomForestRegressor()
rf.fit(X_train,Y_train)

xgb= GradientBoostingRegressor()
xgb.fit(X_train,Y_train)

xg= XGBRegressor()
xg.fit(X_train,Y_train)
```

```python
#prediction on test data
y_pred1=lr.predict(X_test)
y_pred2=las.predict(X_test)
y_pred3=rf.predict(X_test)
y_pred4=xgb.predict(X_test)
y_pred5=xg.predict(X_test)
```

```python
#Evaluating the algorithm                                                                                                                                                                                                                                             
score1=metrics.r2_score(Y_test,y_pred1)
score2=metrics.r2_score(Y_test,y_pred2)
score3=metrics.r2_score(Y_test,y_pred3)
score4=metrics.r2_score(Y_test,y_pred4)
score5=metrics.r2_score(Y_test,y_pred5)
```

print(score1,score2,score3,score4,score5)

```python
print(score1,score2,score3,score4,score5)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
MSE1=mean_squared_error(Y_test,y_pred1)
MSE2=mean_squared_error(Y_test,y_pred2)
MSE3=mean_squared_error(Y_test,y_pred3)
MSE4=mean_squared_error(Y_test,y_pred4)
MSE5=mean_squared_error(Y_test,y_pred5)

MAE1=mean_absolute_error(Y_test,y_pred1)
MAE2=mean_absolute_error(Y_test,y_pred2)
MAE3=mean_absolute_error(Y_test,y_pred3)
MAE4=mean_absolute_error(Y_test,y_pred4)
MAE5=mean_absolute_error(Y_test,y_pred5)
```

```python
print(MSE1,MSE2,MSE3,MSE4,MSE5) # MSE

print(MAE1,MAE2,MAE3,MAE4,MAE5) # MAE
```

```python
final_score=pd.DataFrame({'Models': ['lr','las','rf','xgb','xg'],
"R2_Score" : [score1,score2,score3,score4,score5]})
```
